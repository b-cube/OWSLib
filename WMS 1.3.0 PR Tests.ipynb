````python
%reload_ext autoreload
%autoreload 2

import pprint
from owslib.wms import WebMapService

pp = pprint.PrettyPrinter(indent=4)

# bits from here: http://nbviewer.ipython.org/github/Unidata/unidata-python-workshop/blob/master/wms_sample.ipynb#mainLibs
def loadTmpImage(image_name):
    from IPython.core.display import Image
    from os.path import join
    filename = join('notebooks', 'outputs', image_name)
    return Image(filename)

#Function that saves the layer as an image
def saveLayerAsTmpImage(layer, inname):
    from os.path import join
    full_img_path = join('notebooks', 'outputs', inname)
    out = open(full_img_path, 'wb')
    out.write(layer.read())
    out.close()
````
